```python
#Making a pie chart for top 3 teams with most wins after batting first
import matplotlib.pyplot as plt
import pandas as pd

# Read the IPL dataset
ipl = pd.read_csv(r"C:\Users\DELL\Documents\IPL project\ipl.csv")

# Filter teams that won by runs (batting first)
batting_first = ipl[ipl["Win_By_Runs"] != 0]

# Get the top 3 teams with most wins after batting first
top_teams = batting_first["Winner"].value_counts().head(3)

# Create a pie chart
plt.figure(figsize=(7, 7))
plt.pie(top_teams.values, labels=top_teams.index, autopct="%1.1f%%", startangle=140, colors=["blue", "yellow", "orange"])

# Show the plot
plt.title("Top 3 Teams with Most Wins After Batting First")
plt.show()
```

```python
#Making a histogram for  with most wins after batting first

import matplotlib.pyplot as plt
import pandas as pd

# Read the IPL dataset
ipl = pd.read_csv(r"C:\Users\DELL\Documents\IPL project\ipl.csv")

# Filter teams that won by runs (batting first)
batting_first = ipl[ipl["Win_By_Runs"] != 0]

# Get the count of wins for each team after batting first
win_counts = batting_first["Winner"].value_counts()

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(batting_first["Winner"], bins=len(win_counts), color="blue", edgecolor="black")

# Add titles and labels
plt.title("Histogram of Wins After Batting First")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
